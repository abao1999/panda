```python
import os

from panda.baselines.baselines import (
    FourierARIMABaseline,
    FourierBaseline,
    MeanBaseline,
)
from panda.utils import load_dyst_samples, plot_forecast_1d, safe_standardize
```

```python
work_dir = os.environ["WORK"]
dysts = load_dyst_samples(
    "Laser",
    f"{work_dir}/data/final_base40",
    "train",
    one_dim_target=False,
    num_samples=100,
)
dysts = safe_standardize(dysts, axis=1)
```

```python
batch_context = dysts[:5, :, :512].transpose(0, 2, 1)
batch_ground_truth = dysts[:5, :, 512 : 512 + 128].transpose(0, 2, 1)
```

```python
meanmodel = MeanBaseline(prediction_length=128)
mean_forecasts = meanmodel(batch_context)
plot_forecast_1d(mean_forecasts[0].T, batch_ground_truth[0].T, batch_context[0].T)
```

```python
fftmodel = FourierBaseline(prediction_length=128)
fft_forecasts = fftmodel(batch_context)

plot_forecast_1d(fft_forecasts[0].T, batch_ground_truth[0].T, batch_context[0].T)
```

```python
batched_arimamodel = FourierARIMABaseline(
    prediction_length=128, order=(4, 1, 4), num_fourier_terms=2
)
preds = batched_arimamodel(batch_context)
for i in range(5):
    plot_forecast_1d(preds[i].T, batch_ground_truth[i].T, batch_context[i].T)
```
